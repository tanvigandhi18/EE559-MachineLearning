```python
import numpy as np
import pandas as pd
from sklearn.preprocessing import StandardScaler,PolynomialFeatures
from sklearn.linear_model import Perceptron
from sklearn.model_selection import LearningCurveDisplay
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist
from plotSVMBoundaries import plotSVMBoundaries
#from utils import Polynomial_Mapping, accuracy
from sklearn.metrics import accuracy_score
from prettytable import PrettyTable 
from scipy.special import comb
from sklearn.svm import SVC

```

```python
def openFile(dataset):
    # Read the CSV file using pandas
    data = pd.read_csv(dataset, header=None, skiprows=1)
    
    # Extract features and labels
    X = data.iloc[:, :-1].values.astype(float)
    y = data.iloc[:, -1].values.astype(int)
    
    return X, y

def accuracy(y_pred, y):
    error_count = 0
    for i in range(0,len(y)):
        if y_pred[i] != y[i]:
            error_count+=1
    acc = ((len(y)- error_count)/ len(y)) * 100
    return acc

```

```python
X_train1, y_train1 = openFile("dataset1_train.csv")   #passing it as training, label_train in plot file
X_test1, y_test1 = openFile("dataset1_test.csv")
X_train2, y_train2 = openFile("dataset2_train.csv")   #passing it as training, label_train in plot file
X_test2, y_test2 = openFile("dataset2_test.csv")
X_train3, y_train3 = openFile("dataset3_train.csv")   #passing it as training, label_train in plot file
X_test3, y_test3 = openFile("dataset3_test.csv")
```

Part a)

```python
C_values = [0.01, 1, 100]   #Using a linear kernel, run SVM for 𝐶 = 0.01,  𝐶 = 1,  𝐶 = 100


models = []
for C in C_values:
    model1 = SVC(kernel='linear', C=C)
    model1.fit(X_train1, y_train1)
    y_train_pred1 = model1.predict(X_train1)
    y_test_pred1 = model1.predict(X_test1)
    arr1 = model1.support_vectors_
    print("Train accuracy dataset 1 when C = {} is {}".format(C, accuracy(y_train_pred1, y_train1)))
    print("Test accuracy dataset 1 when C = {} is {}".format(C, accuracy(y_test_pred1, y_test1)))
    print("weight_vector:", model1.coef_)
    print("bias:", model1.intercept_)
    plotSVMBoundaries(X_train1, y_train1, model1, support_vectors = arr1)
    plotSVMBoundaries(X_test1, y_test1, model1)
    
    models.append(model1)
    model2 = SVC(kernel='linear', C=C)
    model2.fit(X_train2, y_train2)
    y_train_pred2 = model2.predict(X_train2)
    y_test_pred2 = model2.predict(X_test2)
    arr2 = model2.support_vectors_
    print("Train accuracy dataset 2 when C = {} is {}".format(C, accuracy(y_train_pred2, y_train2)))
    print("Test accuracy dataset 2 when C = {} is {}".format(C, accuracy(y_test_pred2, y_test2)))
    print("weight_vector:", model2.coef_)
    print("bias:", model2.intercept_)
    plotSVMBoundaries(X_train2, y_train2, model2, support_vectors = arr2)
    plotSVMBoundaries(X_test2, y_test2, model2)
    models.append(model2)
 
```

Part b) #Use RBF kernel, and run SVM for 𝐶 = 1,  𝛾 = 1,  10


```python

C = 1
gamma_values = [1, 10]
models = []

for gamma in gamma_values:
    model1 = SVC(kernel='rbf', C=C, gamma=gamma)
    model1.fit(X_train1, y_train1)
    y_train_pred1 = model1.predict(X_train1)
    y_test_pred1 = model1.predict(X_test1)
    arr1 = model1.support_vectors_
    print("Train accuracy dataset 1 when C = {} and gamma = {} is {}".format(C, gamma, accuracy(y_train_pred1, y_train1)))
    print("Test accuracy dataset 1 when C = {} and gamma = {} is {}".format(C, gamma, accuracy(y_test_pred1, y_test1)))
    plotSVMBoundaries(X_train1, y_train1, model1, support_vectors=arr1)
    plotSVMBoundaries(X_test1, y_test1, model1)
    models.append(model1)
    
    model2 = SVC(kernel='rbf', C=C, gamma=gamma)
    model2.fit(X_train2, y_train2)
    y_train_pred2 = model2.predict(X_train2)
    y_test_pred2 = model2.predict(X_test2)
    arr2 = model2.support_vectors_
    print("Train accuracy dataset 2 when C = {} and gamma = {} is {}".format(C, gamma, accuracy(y_train_pred2, y_train2)))
    print("Test accuracy dataset 2 when C = {} and gamma = {} is {}".format(C, gamma, accuracy(y_test_pred2, y_test2)))
    plotSVMBoundaries(X_train2, y_train2, model2, support_vectors=arr2)
    plotSVMBoundaries(X_test2, y_test2, model2)
    models.append(model2)
    
```
