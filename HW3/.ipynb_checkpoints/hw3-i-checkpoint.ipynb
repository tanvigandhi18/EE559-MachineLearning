```python
# Copyright (C) 2024 Shipeng Liu
# 
# This file is part of EE559 HW3.
# 
# Purpose: Implementation of perceptron algorithms and methods required for EE559 Homework 3.
# Please fill in function perceptron_learning and perceptron_testing
# TA: Shipeng Liu

import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd
from sklearn.preprocessing import normalize
from sklearn.utils import shuffle
```

```python
# functions:
def data(dataframe):
    dataframe = dataframe.to_numpy()
    dataframe = shuffle(dataframe, random_state= 42)
    features = dataframe[:,:-1]
    labels = dataframe[:,-1]
    return features, labels

def perceptron_learning(x, labels, w_int=np.array([1.0,1.0,1.0]),
                        eta=1, max_iterations=1000):
    """
    Implements the batch gd perceptron learning algorithm.

    Args:
        x: (N, D) data array, non-augmented format.
        labels: Array of length N with labels 1, 2.
        w_int: Initial weights as a numpy array. Defaults to numpy.array([1.0,1.0,1.0]).
        eta: Learning rate as a float. Defaults to 1.
        max_iterations: Maximum number of iterations as an int. 
                        Defaults to 1000(here refer to epochs for batch gd).

    Returns:
        w_hats: Weights after each iteration.
                The array size should be (epochs, D+1) when augmented
        Js: Criterion function values after each iteration, size: (epochs, )
        error_rates: Misclassification rates after each iteration, size: (epochs, )
    """
    # x - > features:
    # now for augmentation:
    ones = np.ones((x.shape[0],1))
    
    # now for augmenting the matrix:
    x_augmented = np.column_stack((ones,x))
    var = x_augmented.shape[0]
    # declaring variables for returning Js, new weight values and error_rates
    w_hats = []
    # i will create a copy here as it forms an error recursing over the entire dataset
    # also i need to append the new values in w hats
    w = w_int.copy()
    Js = []
    error_rates = []
    # criterion = []
    
    # training loop for 1000 epochs:
    for epoch in range(max_iterations):
        # small J initially for putting in thr criterion:
        J = 0
        misclassify = 0
        # The reason i use np.zeros_like is if i just consider it as a list I am unable to 
        # broadcast arrays properly....
        criterion = np.zeros_like(w)
        
        # this loop will sum over for g(znxn)<=0
        for i in range(var):
            """instead of creating seperate reflected points augmentation
            I will just assign labels to -1 by masking over the matrix and using 
            the concept of zn sign taught in the class"""
            zn = 1 if labels[i] == 1 else -1 
            
            # checking error i.e g(znxn)< = 0 to misclassied points:
            if zn * np.dot(w.T, x_augmented[i]) <=0:
                misclassify+=1
                criterion += zn * x_augmented[i]
                J+= (-zn * np.dot(w.T, x_augmented[i]))
            
        # Now we use the weight updation rule:
        # w[i+1] = w[i] + n(sum(g(znxn<=0)znxn)
        w = w + (eta * criterion)
        w_hats.append(w)
        Js.append(J)
        error_rates.append(misclassify / var)
        
        # Now halting conditions for breaking the loop:
        # h1 =  all correct classifications 
        # h2 = 1000 iterations reached...
        """ the g(znxn)< = 0 sum are basically being counted when i increment the 
        misclassify variable """
        if misclassify == 0:
            break
        elif epoch == max_iterations:
            break
        
    print(error_rates[-1]) 
    return np.array(w_hats), np.array(Js), np.array(error_rates)

def perceptron_testing(x, labels, w_hat):
    '''
    Implements the perceptron testing.

    Args:
        x: (N, D) data array, non-augmented format.
        labels: Array of length N with labels 1, 2.
        w_hat: final learned weights as a numpy array.
    Returns:
        error_rates: Misclassification rates.
        
    '''
    # x - > features, misclassify is as same as I have used in perceptron learning:
    ones = np.ones((x.shape[0],1))
    x_augmented = np.column_stack((ones,x))
    var = x_augmented.shape[0]
    misclassify = 0
    for i in range(var):
        zn = 1 if labels[i] == 1 else -1 
        if zn * np.dot(w_hat, x_augmented[i]) <= 0:
            misclassify += 1
    
    error_rate = misclassify / var
    print(error_rate)
    return error_rate


def plot_linear_learning(x, labels, w_hat, Js, error_rates):
    """
    Plot training curve, and visualize the decision boundry
    Args:
        x: (N, D) data array, non-augmented format
        labels:  lengnth array with labels 1, 2
        w_hat: final learned weights.
        Js: length N_epochs of Loss values (the learning curve)
        error_rates: length N_epochs of misclassification rates as the algorithm learned
    """
    N_epochs = len(Js)
    x_1 = x[labels==1]
    x_2 = x[labels==2]

    fig, ax = plt.subplots(1, 2, sharex=False, figsize=(12, 6))

    ## scatter plot with deciscion boundaries
    LIMIT = np.max(x)
    x_plot = np.arange(-1 * LIMIT, LIMIT, 0.01)
    ax[0].scatter(x_1.T[0], x_1.T[1], fc=(0, 0, 1, 0.5), label='class 1')
    ax[0].scatter(x_2.T[0], x_2.T[1], fc=(1, 0, 0, 0.5), label='class 2')
    #plot 2-class linear decision boundary
    print("W_hat 1:", w_hat[1])
    ax[0].plot( x_plot, -1 * ( w_hat[1] *  x_plot  + w_hat[0] )
                                     / w_hat[2], linewidth=2, label='boundry')
    ax[0].set_xlabel('x1')
    ax[0].set_ylabel('x2')
    ax[0].set_xlim([-LIMIT, LIMIT])
    ax[0].set_ylim([-LIMIT, LIMIT])
    ax[0].legend()
    ax[0].grid(':')

    ## Learning curve
    epochs = np.arange(N_epochs)
    ax[1].plot(epochs, Js, marker='o', color='green', label='J (Loss)')
    # add second y-axis for the metric of error rate: https://pythonguides.com/matplotlib-two-y-axes/
    ax2 = ax[1].twinx() 
    ax2.plot(epochs, error_rates * 100, marker='x', color='purple', label='Error Rate')
    ax[1].set_ylabel('J (Loss)', color = 'green') 
    ax[1].tick_params(axis ='y', labelcolor = 'green') 
    ax2.set_ylabel('Error rate (%)', color = 'purple') 
    ax2.tick_params(axis ='y', labelcolor = 'purple') 

    # ax[1].set_xlabel('epoch')
    # ax[1].set_ylabel('J (Loss)')
    ax2.set_ylabel('error rate (%)')
    ax[0].legend()
    ax[1].grid(':')
    
def plot_linear_testing(x, labels, w_hat):
    """
    Visualize the decision boundary of testing sets
    Args:
    x: (N, D) data array, non-augmented format
    labels:  lengnth array with labels 1, 2
    w_hat: final learned weights as a numpy array.
    """
  
    x_1 = x[labels==1]
    x_2 = x[labels==2]

    fig, ax = plt.subplots(1, 2, sharex=False, figsize=(12, 6))

    ## scatter plot with deciscion boundaries
    
    LIMIT = np.max(x)

    
    x_plot = np.arange(-1 * LIMIT, LIMIT, 0.01)
    ax[0].scatter(x_1.T[0], x_1.T[1], fc=(0, 0, 1, 0.5), label='class 1')
    ax[0].scatter(x_2.T[0], x_2.T[1], fc=(1, 0, 0, 0.5), label='class 2')
    #plot 2-class linear decision boundary
    ax[0].plot( x_plot, -1 * ( w_hat[1] *  x_plot  + w_hat[0] ) 
                                    / w_hat[2], linewidth=2, label='boundry')
    ax[0].set_xlabel('x1')
    ax[0].set_ylabel('x2')
    ax[0].set_xlim([-LIMIT, LIMIT])
    ax[0].set_ylim([-LIMIT, LIMIT])
    ax[0].legend()
    ax[0].grid(':')
```

```python
df1 = pd.read_csv("dataset1_train.csv")
df1_t = pd.read_csv("dataset1_test.csv")
df2 = pd.read_csv("dataset2_train.csv")
df2_t = pd.read_csv("dataset2_test.csv")
df3 = pd.read_csv("dataset3_train.csv")
df3_t = pd.read_csv("dataset3_test.csv")
```

```python
features1, labels1 = data(df1)
w_hats1, Js, error_rates = perceptron_learning(features1, labels1)
plot_linear_learning(features1, labels1, w_hats1[-1], Js, error_rates)
plt.show()
```

```python
features1_t, labels1_t = data(df1_t)
error_rate1 = perceptron_testing(features1_t, labels1_t,w_hats1[-1])
plot_linear_testing(features1_t, labels1_t, w_hats1[-1])
plt.show()
```
