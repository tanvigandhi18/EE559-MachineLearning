```python
import numpy as np
from sklearn.linear_model import Perceptron
from sklearn.model_selection import KFold
from sklearn.metrics import accuracy_score

from matplotlib.colors import ListedColormap

# Using the plotting function from HW1

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist

# Using the plotting function from HW1

import numpy as np
import matplotlib.pyplot as plt
from scipy.spatial.distance import cdist


def plotDecArea(X_original, y_original, model, tag):
    """
    plot the Decision Area
    Args:
        X_train_original (ndarray (n,2)): original data, n examples with 2 features
        y_original (ndarray (n,)) : target values
        model: sklearn Perceptron
        augmentation_method: preprocessing method (polynomial transformation, scaler (if there is)
    """

    training_x = X_original
    training_y = y_original

    # Set the feature range for ploting
    max_x = np.ceil(max(training_x[:, 0])) + 1
    min_x = np.floor(min(training_x[:, 0])) - 1
    max_y = np.ceil(max(training_x[:, 1])) + 1
    min_y = np.floor(min(training_x[:, 1])) - 1

    xrange = (min_x, max_x)
    yrange = (min_y, max_y)

    # step size for how finely you want to visualize the decision boundary.
    inc = 0.005

    # generate grid coordinates. this will be the basis of the decision
    # boundary visualization.
    (x, y) = np.meshgrid(np.arange(xrange[0], xrange[1]+inc/100, inc), np.arange(yrange[0], yrange[1]+inc/100, inc))

    # size of the (x, y) image, which will also be the size of the
    # decision boundary image that is used as the plot background.
    image_size = x.shape
    xy = np.hstack( (x.reshape(x.shape[0]*x.shape[1], 1, order='F'), y.reshape(y.shape[0]*y.shape[1], 1, order='F')) ) # make (x,y) pairs as a bunch of row vectors.

    # distance measure evaluations for each (x,y) pair.
    pred_label = model.predict(xy)
        
    # reshape the idx (which contains the class label) into an image.
    decisionmap = pred_label.reshape(image_size, order='F')

    #show the image, give each coordinate a color according to its class label
    plt.imshow(decisionmap, extent=[xrange[0], xrange[1], yrange[0], yrange[1]], cmap='cividis', origin='lower', aspect='auto')
    
    plt.scatter(training_x[training_y==1][:, 0], training_x[training_y==1][:, 1], marker='d', label = 'Class 1')
    plt.scatter(training_x[training_y==2][:, 0], training_x[training_y==2][:, 1], marker = 'o', label = 'Class 2')
    plt.scatter(training_x[training_y==3][:, 0], training_x[training_y==3][:, 1], marker = '+', label = 'Class 3')
    # for label in np.unique(training_y):
    #     plt.scatter(training_x[training_y == label, 0], training_x[training_y == label, 1], label=f'Class {int(label)}')

    plt.xlim(xrange)
    plt.ylim(yrange)
    plt.title(f"Run with the {tag} classification error rate")
    plt.legend()
    plt.show()
```

```python
def run_perceptron(space, labels, n_splits=20, n_runs=5):
    total_errors = []
    models_after_1st_fold = []

    for run in range(n_runs):
        kf = KFold(n_splits=n_splits, shuffle=True, random_state=run)
        shuffled_features, shuffled_labels = shuffle(space, labels, random_state=run)
        run_errors = []

        for fold_index, (train_index, val_index) in enumerate(kf.split(shuffled_features)):
            train_data, val_data = shuffled_features[train_index], shuffled_features[val_index]
            train_labels, val_labels = shuffled_labels[train_index], shuffled_labels[val_index]
            model = Perceptron()
            model.fit(train_data, train_labels)
            predicted_labels = model.predict(val_data)
            local_error = 1 - accuracy_score(val_labels, predicted_labels)
            run_errors.append(local_error)

            if run == 0 and fold_index == 0:  # Store the first model from the first fold of the first run
                first_fold_model = model

        total_errors.append(np.mean(run_errors))
        models_after_1st_fold.append(first_fold_model)
        print(f"Average of Mean Classification Error for Run {run + 1}: {total_errors[-1]:.4f}")
        print(f"Standard Deviation for Run {run + 1}: {np.std(run_errors):.4f}")

    average_error = np.mean(total_errors)
    std_dev_error = np.std(total_errors)
    print(f"Average of Mean Classification Error for {n_splits} folds: {average_error:.4f}")
    print(f"Standard Deviation of Mean Classification Error for {n_splits} folds: {std_dev_error:.4f}")

    lowest_error_model = models_after_1st_fold[np.argmin(total_errors)]
    highest_error_model = models_after_1st_fold[np.argmax(total_errors)]
    plotDecArea(space, labels, model=lowest_error_model, tag='lowest')
    plotDecArea(space, labels, model=highest_error_model, tag='highest')
    #plotting(space, models_after_1st_fold, total_errors, mode = 'low')
    #plotting(space, models_after_1st_fold, total_errors, mode = 'high')

    return models_after_1st_fold, total_errors, std_dev_error

# Assuming 'features' and 'labels' are defined
space = features[:, [1, 2]]  # Example feature columns
weights, total_errors, std_errors = run_perceptron(space, labels)
```
